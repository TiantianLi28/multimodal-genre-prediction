```python
import pandas as pd
from fuzzywuzzy import fuzz
import langid
```

```python
data = pd.read_csv('scraped_israel.csv', header=None)
```

```python
track_columns=['track_id','track_name','track_artist','lyrics','track_popularity','track_album_id','track_album_name','track_album_release_date','playlist_name','playlist_id','playlist_genre','playlist_subgenre','language', 'track_id','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
data.columns = track_columns
```

```python
data.dropna(subset='lyrics', inplace=True)
```

```python
data['lyrics'] = data['lyrics'].apply(lambda x: x.lower())
data['track_name'] = data['track_name'].apply(lambda x: x.lower() if not pd.isna(x) else x)
```

```python
data.drop_duplicates(subset='lyrics', inplace=True)
```

```python
data.drop_duplicates(subset='track_name', inplace=True)
```

```python
# Function to detect language using langid
def detect_language(text):
    lang, _ = langid.classify(text)
    return lang

# Apply language detection to the 'text' column
data['language'] = data['lyrics'].apply(detect_language)

# Filter out non-English entries
data = data[data['language'] == 'en'].drop(columns=['language'])
```

```python
i = 0
data = data.sort_values(by='track_name')
to_remove = []
while i < data.shape[0]:
    name_list = data.iloc[i,1].split()
    name_length = len(name_list)
    for j in range(1,20):
        if i+j<data.shape[0]:
            name_j_list = data.iloc[i+j, 1].split()
            name_j_length = len(name_j_list)
            if name_j_list[:min(3, name_j_length)] == name_list[:min(3, name_length)]:
                to_remove.append(i+j)
            else:
                break
    i+=1
        
```
